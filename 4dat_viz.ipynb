```python
# import os
import numpy as np
import matplotlib.pyplot as plt
import struct
import tkinter as tk
from tkinter import filedialog
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']  
plt.rcParams['axes.unicode_minus'] = False  
```

```python
def ad2mv(b):
    value = struct.unpack("h", b)
    return value[0]
```

```python
def open_file():

    file_path = filedialog.askopenfilename(filetypes=[("DAT files", "*.dat")])
    f = open(file_path, "rb") 
    print(file_path)
    # Do something with the file_path, such as opening the file

    dat_leads = ['I', 'II', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6']
    ecg_dict = {}

    for i in range(8):
        lead_data = []

        for pos in range(0, 500 * 10):
            b = f.read(2)
            value = ad2mv(b)
            lead_data.append(value)

        ecg_dict[dat_leads[i]] = np.array(lead_data)
        print(lead_data)
    # III = II-I
    # aVR =  -0.5*(I+II)
    # aVL = I - 0.5*II
    # aVF = II - 0.5*I
    ecg_dict['III'] = ecg_dict['II'] - ecg_dict['I']
    ecg_dict['aVR'] = -0.5 * (ecg_dict['I'] + ecg_dict['II'])
    ecg_dict['aVL'] = ecg_dict['I'] - 0.5 * ecg_dict['II']
    ecg_dict['aVF'] = ecg_dict['II'] - 0.5 * ecg_dict['I']

    f.close()

    print('I:', ecg_dict['I'])
    print('II:', ecg_dict['II'])
    print('III:', ecg_dict['III'])
    print('aVR:', ecg_dict['aVR'])
    twelve_lead = ['I', 'II', 'III', 'aVR', 'aVL', 'aVF', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6']
    fig, axs = plt.subplots(nrows=12, sharex=True, figsize=(12, 8))
    for i in range(12):
        axs[i].plot(ecg_dict[twelve_lead[i]])
        axs[i].set_ylabel(twelve_lead[i])
    filename = file_path.split('/')[-1].replace('.dat', '')
    # plt.xlabel(f'FileName: {filename}')
    plt.suptitle(f'FileName: {filename}')
    plt.subplots_adjust(top=0.92)
    plt.show()
```

```python
if __name__ == '__main__':
    root = tk.Tk()
    button = tk.Button(root, text="Open DAT file", command=open_file)
    button.pack()

    root.mainloop()
```
